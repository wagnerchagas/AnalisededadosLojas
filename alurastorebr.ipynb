### Importação dos dados

# 2. Vendas por Categoria


```python
def analisar_vendas_por_categoria(df, nome_loja):
    vendas = df.groupby("Categoria do Produto")["Valor Final"].sum().reset_index()
    vendas = vendas.sort_values(by="Valor Final", ascending=False)
    print(f"\nVendas por categoria - {nome_loja}:\n")
    print(vendas)
```

# 3. Média de Avaliação das Lojas

```python
Avaliacao = {"Loja": 4.5, "Loja2": 3.8, "Loja3": 4.9}

for loja, media in Avaliacao.items():
    print(f"{loja}: {media:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
todos = pd.concat([loja, loja2, loja3, loja4])
mais_vendido = todos['Produto'].value_counts().idxmax()
menos_vendido = todos['Produto'].value_counts().idxmin()

print("Produto mais vendido:", mais_vendido)
print("Produto menos vendido:", menos_vendido)
```

# 5. Frete Médio por Loja


```python
frete_medio = {
    'Loja ': loja['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean(),
}

for loja, frete in frete_medio.items():
    print(f"{loja}: R$ {frete:.2f}")
```
